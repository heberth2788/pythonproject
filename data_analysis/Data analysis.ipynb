## Custom example

```python
anhos = [2018, 2019, 2020, 2021, 2022, 2023]
tempPe = [32, 18, 30, 29, 34, 20]
tempCh = [30, 22, 29, 32, 30, 20]

plt.figure(figsize=(10, 4))

plt.plot(anhos, tempPe, linestyle="-", marker="o", color="b", label="PE")
plt.plot(anhos, tempCh, linestyle="-", marker="o", color="r", label="CH")

plt.xlabel("Years")
plt.ylabel("Temps")
plt.title("Years & Temps")
plt.legend()
plt.savefig("years_vs_temps.jpg")

plt.show()
```

## Sub plots

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
ax[0].plot(anhos,tempPe, color='g', label="PE")
ax[0].legend()
ax[1].plot(anhos,tempCh, color='b', label="CH")
ax[1].legend()
plt.show()
```

## Bar chart

```python
countries = ["PE", "AR", "CH", "BR", "UY", "CO"]
cups = [2, 3, 2, 9, 4, 5]
plt.figure(figsize = (10, 4))
plt.bar(countries, cups, color = "b")
plt.xlabel("Country")
plt.ylabel("No")
plt.title("Countries and their cups")

plt.show()
```

# Pie chart

```python
plt.pie(cups, labels = countries, autopct = "%1.1f%%")
plt.title("Cups by Country")
plt.savefig("cups_by_country.jpg")
plt.legend()
plt.show()
```

## Histogram chart

```python
ages = [10, 18, 30, 28, 29, 30, 19, 5, 2, 18, 20, 30, 32, 39, 10, 5, 6, 80]
bins = [0, 15, 20, 21, 30, 40]
plt.hist(ages, bins, edgecolor = "red")
plt.xlabel("Ages")
plt.ylabel("Qty")
plt.title("Ages histogram")
plt.savefig("ages_histogram.jpg")
plt.show()
```

## Box chart

```python
plt.boxplot(ages)
plt.show()
```
